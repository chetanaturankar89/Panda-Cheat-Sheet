# PANDAS

-  Pandas is an open source data analysis library written in python.
-  It leverages the power and speed of numpy to make data analysis and preprocessing easy for data scientists.
-  It provides rich and highly robust data operations

```python
import pandas as pd
```

## Import Export Data

```python
# Read data from CSV File 
sales = pd.read_csv('sales.csv')
sales
```

Importing data using below 
1. pd.read_csv(filename): Read data from a CSV file.
2. pd.read_table(filename): Read data from a delimited text file.
3. pd.read_excel(filename): Read data from an Excel file.
4. pd.read_sql(query, connection_object): Read data from a SQL table/database.
5. pd.read_json(json_string): Read data from a JSON formatted string, URL, or file.
6. pd.read_html(url): Parse an HTML URL, string, or file to extract tables to a list of DataFrames.
7. pd.DataFrame(dict): Create a DataFrame from a dictionary (keys as column names, values as lists).
8. df.to_csv(filename): Write to a CSV file.
9. df.to_excel(filename): Write to an Excel file.
10. df.to_sql(table_nm, connection_object): Write to a SQL table.
11. df.to_json(filename): Write to a file in JSON format.
